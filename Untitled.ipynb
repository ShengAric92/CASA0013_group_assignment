```python
import requests

# URL of the file to be downloaded
url = "http://data.insideairbnb.com/united-kingdom/england/london/2023-09-06/visualisations/listings.csv"

# Sending a GET request to the URL
response = requests.get(url)

# Checking if the request was successful
if response.status_code == 200:
    # Writing the content of the response to a file
    with open('listings.csv', 'wb') as file:
        file.write(response.content)
    message = "File downloaded successfully."
else:
    message = "Failed to download the file. Status code: {}".format(response.status_code)

message
```
